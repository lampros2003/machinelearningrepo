# Problem 1: α)

### Απάντηση

Από την εκφώνηση έχουμε τα σημαντικά δεδομένα:

- **Α)** Τα $( x_1, x_2 )$ είναι ανεξάρτητα, άρα $f(x_1, x_2) = f(x_1) f(x_2) $.
- **Β)** Οι δύο υποθέσεις έχουν ίσα priors, $ P(H_0) = P(H_1) = 0.5 $.

Έτσι, δημιουργείται ο παρακάτω πίνακας σύμφωνα με τον Bayes:

|                         | $H_0 $      | $H_1 $      |
| ----------------------- | ----------- | ----------- |
| Prior Probability       | 0.5         | 0.5         |
| P. D.F.                 | $ f_0(x) $  | $ f_1(x) $  |
| Correct Decision Cost   | $ C_{00} $ | $ C_{11} $ |
| Incorrect Decision Cost | $ C_{01} $ | $ C_{10} $ |

### Optimum Bayes Test

Σύμφωνα με τη θεωρία (Lecture 5, pg 8), το βέλτιστο τεστ Bayes είναι το παρακάτω likelihood ratio test:

$$
H_1 \text{, if } \frac{f_1(x_1, x_2)}{f_0(x_1, x_2)} > \frac{P(H_0) C_{00} - P(H_0) C_{01}}{P(H_1) C_{10} - P(H_1) C_{11}}
$$

$$
H_0 \text{, if } \frac{f_1(x_1, x_2)}{f_0(x_1, x_2)} < \frac{P(H_0) C_{00} - P(H_0) C_{01}}{P(H_1) C_{10} - P(H_1) C_{11}}
$$

Λόγω των (Α) και (Β), το τεστ γίνεται:

$$
H_1 \text{, if } \frac{f_1(x_1) \cdot f_1(x_2)}{f_0(x_1) \cdot f_0(x_2)} > \frac{C_{00} - C_{01}}{C_{10} - C_{11}}
$$

$$
H_0 \text{, if } \frac{f_1(x_1) \cdot f_1(x_2)}{f_0(x_1) \cdot f_0(x_2)} < \frac{C_{00} - C_{01}}{C_{10} - C_{11}}
$$

### Ισότητα Μέσου Κόστους και Πιθανότητας Σφάλματος

Γνωρίζουμε ότι το average cost ισούται με την error probability όταν:

$$
\frac{C_{00} - C_{01}}{C_{10} - C_{11}} = 1
$$

(Lecture 5, pg 8)

Σε αυτήν την περίπτωση, ο πίνακας έχει τη μορφή:

|                         | $H_0 $     | $H_1 $     |
| ----------------------- | ---------- | ---------- |
| Prior Probability       | 0.5        | 0.5        |
| P. D.F.                 | $ f_0(x) $ | $ f_1(x) $ |
| Correct Decision Cost   | $0 $       | $ 0 $      |
| Incorrect Decision Cost | $ 1 $      | $ 1 $      |

Δηλαδή έχω ίσο μοναδιαίο κόστος για κάθε λάθος και 0 κόστος για σωστή απάντηση

Η τελική μορφή του τεστ λοιπόν είναι:

$$
H_1 \text{, if } \frac{f_1(x_1) \cdot f_1(x_2)}{f_0(x_1) \cdot f_0(x_2)} > 1
$$

$$
H_0 \text{, if } \frac{f_1(x_1) \cdot f_1(x_2)}{f_0(x_1) \cdot f_0(x_2)} < 1
$$

### Εκφράσεις για τις Κατανομές

- $ H_0: f_0(x) \sim N(0, 1) $, άρα:

$$
f_0(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

- $H_1: f_1(x) \sim 0.5[N(-1, 1) + N(1, 1)] $, άρα:

$$
f_1(x) = \frac{1}{2} \frac{1}{\sqrt{2\pi}} \left(e^{-\frac{(x+1)^2}{2}} + e^{-\frac{(x-1)^2}{2}}\right)
$$

### Λόγος Πιθανοφάνειας

$$
\frac{f_1(x)}{f_0(x)} = \frac{\frac{1}{2} \left(e^{-\frac{(x+1)^2}{2}} + e^{-\frac{(x-1)^2}{2}}\right)}{e^{-\frac{x^2}{2}}}
$$

Με την ανεξαρτησία των $ x_1, x_2 $ (Α):

$$
\frac{f_1(x_1, x_2)}{f_0(x_1, x_2)} = \frac{f_1(x_1)}{f_0(x_1)} \cdot \frac{f_1(x_2)}{f_0(x_2)}
$$

Η τελική σχέση του τεστ είναι:

$$
\frac{1}{2} \left(e^{-\frac{(x_1 + 1)^2}{2}} + e^{-\frac{(x_1 - 1)^2}{2}}\right) \cdot \frac{1}{2} \left(e^{-\frac{(x_2 + 1)^2}{2}} + e^{-\frac{(x_2 - 1)^2}{2}}\right) > e^{-\frac{x_1^2}{2}} e^{-\frac{x_2^2}{2}}
$$

### Απλοποίηση

Απλοποιώντας:

$$
e^{-\frac{(x_i \pm 1)^2}{2}} = e^{-\frac{x_i^2}{2}} \cdot e^{\mp x_i} \cdot e^{-\frac{1}{2}}
$$

Άρα:

$$
\frac{1}{4} \left(e^{-\frac{x_1^2}{2}}e^{-\frac{1}{2}}(e^{x_1} + e^{-x_1})\right) \left(e^{-\frac{x_2^2}{2}}e^{-\frac{1}{2}}(e^{x_2} + e^{-x_2})\right) > e^{-\frac{x_1^2}{2}} e^{-\frac{x_2^2}{2}}
$$

Γνωρίζοντας ότι:

$$
e^{x} + e^{-x} = 2 \cosh(x)
$$

Η τελική μορφή είναι:

$$
H_1 \text{, if } \cosh(x_1) \cosh(x_2) > e
$$

$$
H_0 \text{, if } \cosh(x_1) \cosh(x_2) < e
$$


# Problem 1: b)

### Απάντηση

Θέλω να υλοποιήσω, σε python πρόγραμμα μια προσομοίωση της παραπάνω διαδικασίας για τον υπολογισμό error probability.
Αρχικά κάνω import τη βιβλιοθήκη numpy και την συνάρτηση norm από την schipy.stats.
Με την numpy θα μπορώ να κάνω περίπλοκους αριθμητικούς υπολογισμούς γρήγορα, και με την norm θα μπορώ να δημιουργήσω samples από normal distribution.


```python
import numpy as np
from scipy.stats import norm
```

Ορίζω τις συναρτήσεις πυκνότητας των δύο συναρτήσεων όπως τις υπολόγισα παραπάνω.\
Επίσης δημιουργώ την συνάρτηση που προσομοιώνει το Bayes test με δύο τρόπους, by default με την απλοποιημένη σχέση
με τα υπερβολικά συνημίτονα η αλλιώς άμεσα με τις συναρτήσεις. \
Σημαντικό ότι λογαρίθμιζω τα ratios για να μπορώ να κάνω πιο ευσταθείς και ακριβείς υπολογισμούς.


```python
def f0(x):
    """Probability density function for H0"""
    return 1 / np.sqrt(2 * np.pi) * np.exp(-(x**2) / 2)


def f1(x):
    """Probability density function for H1"""
    return 0.5 * (norm.pdf(x, -1, 1) + norm.pdf(x, 1, 1))


def bayes_test(x1, x2, simplified="simple"):
    """Optimal Bayes test"""
    if simplified == "simple":
        log_likelihood_ratio = np.log(np.cosh(x1) * np.cosh(x2))
        if log_likelihood_ratio > np.log(np.e):
            return 1  # Decide H1
        else:
            return 0  # Decide H0

    log_likelihood_ratio = np.log(f1(x1) / f0(x1)) + np.log(f1(x2) / f0(x2))
    if log_likelihood_ratio > 0:
        return 1  # Decide H1
    else:
        return 0  # Decide H0
```

```python
def generate_data(n_samples, hypothesis):
    """Generate data for H0 or H1"""
    if hypothesis == 0:
        # H0: N(0,1)
        return np.random.normal(0, 1, (n_samples, 2))
    else:
        # H1: 0.5{N(-1,1) + N(1,1)}
        means = np.random.choice([-1, 1], size=(n_samples, 2))
        return np.random.normal(means, 1)
```

Δημιουργώ $10^6$ ζεύγη samples για κάθε υπόθεση.\
Εφαρμόζω το bayes test στην απλοποιημένη του μορφή σε κάθε ένα απο αυτά τα ζεύγη samples, αποθηκεύοντας τα αποτελέσματα
στις λίστες y0 για $H_0$ y1 για $H_1$.\
Μετρώ μετά τα λάθη που έχει κάνει το τεστ και τέλος βρίσκω το μέσω σφάλμα.\


```python
num_samples = 10**6


# Generate samples from H0
X_test_h0 = generate_data(1000000, hypothesis=0)
X_test_h1 = generate_data(1000000, hypothesis=1)

# Apply the Bayes test
y0 = [bayes_test(X_test_h0[i][0], X_test_h0[i][1]) for i in range(num_samples)]
y1 = [bayes_test(X_test_h1[i][0], X_test_h1[i][1]) for i in range(num_samples)]

# Calculate error probabilities
error_p0 = np.sum(y0) / num_samples
error_p1 = 1 - np.sum(y1) / num_samples
total_error = 0.5 * error_p0 + 0.5 * error_p1
```

Τυπώνω τα αποτελέσματα.


```python
print(f"Error probability for H0: {error_p0:.4f}")
print(f"Error probability for H1: {error_p1:.4f}")
print(f"Total error probability: {total_error:.4f}")
```

# Problem 1: c)

### Απάντηση

Αρχικά, σχεδιάζω διάφορες βοηθητικές συναρτήσεις:

- **`generate_data`**: Δημιουργεί τα samples.
- **`initialize_parameters`**: Δημιουργεί ένα neural network με τις δωθείσες προδιαγραφές (`input_size`, `hidden_size`, `output_size`), με βάρη \( W_1, W_2 \) αρχικά ορισμένα, όπως είπαμε στην υποσημείωση, και bias \( 0 \).
- **`forward_pass`**: Εκτελεί το forward pass από το neural network.

Για το hidden layer, χρησιμοποίησα τη συνάρτηση ενεργοποίησης **tanh**, καθώς έβγαζα τα καλύτερα αποτελέσματα. Οι συνάρτησεις **ReLU** και **sigmoid** παρουσίαζαν προβλήματα.


```python
# Helper functions
def sigmoid(x):

    return 1 / (1 + np.exp(-x))


def generate_data(n_samples, hypothesis):
    """Generate  for H0 or H1"""
    if hypothesis == 0:
        # H0: N(0,1)
        return np.random.normal(0, 1, (n_samples, 2))
    else:
        # H1: 0.5{N(-1,1) + N(1,1)}
        means = np.random.choice([-1, 1], size=(n_samples, 2))
        return np.random.normal(means, 1)


def initialize_parameters(input_size, hidden_size, output_size):
    W1 = np.random.normal(
        0, np.sqrt(1 / (input_size + hidden_size)), (input_size, hidden_size)
    )
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.normal(
        0, np.sqrt(1 / (hidden_size + output_size)), (hidden_size, output_size)
    )
    b2 = np.zeros((1, output_size))
    return {"W1": W1, "b1": b1, "W2": W2, "b2": b2}


def forward_pass(X, parameters):
    W1, b1 = parameters["W1"], parameters["b1"]
    W2, b2 = parameters["W2"], parameters["b2"]
    # First layer
    Z1 = np.dot(X, W1) + b1
    A1 = np.tanh(Z1)  # Using tanh as activation for hidden layer
    # Output layer
    Z2 = np.dot(A1, W2) + b2
    cache = {"Z1": Z1, "A1": A1, "Z2": Z2}
    return cache
```

Σχεδιάζω ένα class που υλοποιεί τον ADAM optimizer. Για την υλοποίηση του ADAM σε Python, εκτός από τις σημειώσεις στο `derivatives.pdf`, χρησιμοποίησα και το παρακάτω tutorial:
(https://www.geeksforgeeks.org/how-to-implement-adam-gradient-descent-from-scratch-using-python/)

Ο ADAM παρακολουθεί την πρώτη και δεύτερη ροπή (μέση τιμή του slope και ρυθμός αλλαγής του slope) και τροποποιεί τόσο τα biases όσο και τις παραμέτρους του δικτύου. Αυτό επιτρέπει μια ομαλή σύγκλιση κατά τη διάρκεια του training.


```python
class ADAMer:
    def __init__(self, learning_rate=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.m = {}
        self.v = {}
        self.t = 0

    def initialize(self, parameters):
        for key in parameters.keys():
            self.m[key] = np.zeros_like(parameters[key])
            self.v[key] = np.zeros_like(parameters[key])

    def update(self, parameters, gradients):
        self.t += 1

        for key in parameters.keys():
            # Update biased first moment estimate
            self.m[key] = self.beta1 * self.m[key] + (1 - self.beta1) * gradients[key]

            # Update biased second raw moment estimate
            self.v[key] = self.beta2 * self.v[key] + (1 - self.beta2) * np.square(
                gradients[key]
            )

            # Compute bias-corrected first moment estimate
            m_hat = self.m[key] / (1 - self.beta1**self.t)

            # Compute bias-corrected second raw moment estimate
            v_hat = self.v[key] / (1 - self.beta2**self.t)
            # Update parameters
            parameters[key] -= (
                self.learning_rate * m_hat / (np.sqrt(v_hat) + self.epsilon)
            )
```

Οι πίνακες αυτοί αποθηκεύουν τα στοιχεία του training. Δημιουργώ τις βοηθητικές συναρτήσεις `compute_cost_and_error` για τον υπολογισμό των κόστων και των λαθών, και `get_smoothed_metric` για smoothing των δεδομένων.

## Υπολογισμός Κόστους και Λαθών

Υπολογίζω το error και το κόστος με τους γνωστούς τύπους.  
### Υπολογισμός ERROR
Για το cross-entropy:
$$
H_1 \text{ if } \frac{r(x_1, x_2)}{1 + r(x_1, x_2)} > 0.5
$$

$$
H_0 \text{ if } \frac{r(x_1, x_2)}{1 + r(x_1, x_2)} \leq 0.5
$$
και για το exponential:
$$

H_1 \text{ if } \log r(x_1, x_2) > 0
$$


$$

H_0 \text{ if } \log r(x_1, x_2) \leq 0
$$
### Υπολογισμός Κόστους

Βρήκα έναν εύκολο τρόπο υπολογισμού του κόστους σε Python [εδώ](https://www.geeksforgeeks.org/binary-cross-entropy-log-loss-for-binary-classification/?ref=oin_asr1).
Τα κόστη είναι τα εξής(προσαρμοσμένα για κάθε υπόθεση)

1. $$dZ2_{h0} = 0.5 \cdot e^{0.5 \cdot \text{output}_{h0}} $$

2. $$ dZ2_{h1} = -0.5 \cdot e^{-0.5 \cdot \text{output}*{h1}} $$

3. $$ dZ2_{h0} = \frac{\text{output}*{h0}}{1 - \text{output}\_{h0} } $$

4. $$ dZ2_{h1} = -\frac{1}{\text{output}*{h1} } $$

## Training Loop

Για κάθε epoch:

1. Κάνω training στα δύο shuffled datasets, κάνοντας αρχικά ένα forward pass για να πάρω τα δύο layers και το activated layer.
2. Με αυτά, υπολογίζω τα κοστη ανάλογα με τη μέθοδο που χρησιμοποιώ στο network (προσέχω στην περίπτωση cross-entropy να εφαρμόσω το sigmoid στην έξοδο) για κάθε υπόθεση.
3. Στη συνέχεια, ενημερώνω τα gradients δύο φορές – μία για τα δεδομένα κάθε υπόθεσης.

### Σημείωση

Επεξεργάζομαι κάθε υπόθεση ξεχωριστά πάνω στα ίδια αρχικά parameters και ενημερώνω τα parameters και για τις δύο υποθέσεις, κάτι που είναι πρακτικά το ίδιο με το να τα είχα κάνει random mix. Έτσι, κατάφερα να λύσω διάφορα προβλήματα που είχα με το NumPy. Ουσιαστικά κάνω train και με ένα sample απο κάθε υπόθεση, του οποίου θα γρνωρίζω το label.
Αυτό το έκανα διότι όταν έκανα labeling και προσπαθούσα να κάνω mix το numpy παρουσίαζε συνεχώς προβλήματα.

Υπέθεσα οτι δεν θα υπάρχει μεγάλη διαφορά στην απόδωση κάτι το οποίο είδα ισχύει στο τέλος

## Χρήση ADAM Optimizer

Τέλος, καλώ τον ADAM optimizer για να τροποποιήσει τις παραμέτρους. Η συνάρτηση training επιστρέφει το τελικό network με τα errors που παρουσιάζονται κατά το training όταν έχω convergence, ή όταν τα max epochs που όρισα περάσουν.

Τελικά κάνω smoothing τα κόστη και errors των iterations και τα αποθηκεύω.
Για το convergence χρησιμοποίησα τη μέθοδο που μας δόθηκε στην εκφώνηση, με `convergence_threshold = 1e-4`.


```python
def train_network(
    X_train_h0,
    X_train_h1,
    method="cross_entropy",
    epochs=100,
    convergence_threshold=1e-4,
):

    input_size = 2
    hidden_size = 20
    output_size = 1

    # Initialize parameters
    parameters = initialize_parameters(input_size, hidden_size, output_size)
    optimizer = ADAMer(learning_rate=0.001)
    optimizer.initialize(parameters)

    n_samples = len(X_train_h0)

    # Track both costs and errors
    cost_history = []
    error_history = []
    smoothed_costs = []
    smoothed_errors = []

    def compute_cost_and_error(x0_sample, x1_sample, params):
        cache_h0 = forward_pass(x0_sample, params)
        cache_h1 = forward_pass(x1_sample, params)
        ##### Evaluation of errors, costs
        if method == "cross_entropy":
            output_h0 = sigmoid(cache_h0["Z2"])
            output_h1 = sigmoid(cache_h1["Z2"])

            cost_h0 = -np.log(1 - output_h0 + 1e-10)
            cost_h1 = -np.log(output_h1 + 1e-10)
            # Error for cross-entropy
            error_h0 = (output_h0 >= 0.5).mean()
            error_h1 = (output_h1 < 0.5).mean()
        else:  # exponential method
            output_h0 = cache_h0["Z2"]
            output_h1 = cache_h1["Z2"]
            cost_h0 = np.exp(0.5 * output_h0)
            cost_h1 = np.exp(-0.5 * output_h1)

            # Error  for exponential
            error_h0 = (output_h0 >= 0).mean()
            error_h1 = (output_h1 < 0).mean()

        total_cost = np.mean(cost_h0) + np.mean(
            cost_h1
        )  # Average cost across both hypotheses
        total_error = 0.5 * (
            error_h0 + error_h1
        )  # Average error rate across both hypotheses

        return total_cost, total_error
        ##############

    #####
    def get_smoothed_metric(values):
        window_size = min(20, len(values))
        return np.mean(values[-window_size:])

    for epoch in range(epochs):
        # Shuffle training data
        indices = np.random.permutation(n_samples)
        X_h0_shuffled = X_train_h0[indices]
        X_h1_shuffled = X_train_h1[indices]

        epoch_costs = []
        epoch_errors = []

        for i in range(n_samples):
            X_h0_sample = X_h0_shuffled[i : i + 1]
            X_h1_sample = X_h1_shuffled[i : i + 1]

            # Forward pass
            cache_h0 = forward_pass(X_h0_sample, parameters)
            cache_h1 = forward_pass(X_h1_sample, parameters)

            # Compute gradients based on method
            if method == "cross_entropy":
                output_h0 = sigmoid(cache_h0["Z2"])
                output_h1 = sigmoid(cache_h1["Z2"])

                dZ2_h0 = output_h0 / (1 - output_h0 + 1e-10)
                dZ2_h1 = -1 / (output_h1 + 1e-10)
            else:
                output_h0 = cache_h0["Z2"]
                output_h1 = cache_h1["Z2"]
                dZ2_h0 = 0.5 * np.exp(0.5 * output_h0)
                dZ2_h1 = -0.5 * np.exp(-0.5 * output_h1)

            # Initialize gradients
            gradients = {
                "W1": np.zeros_like(parameters["W1"]),
                "b1": np.zeros_like(parameters["b1"]),
                "W2": np.zeros_like(parameters["W2"]),
                "b2": np.zeros_like(parameters["b2"]),
            }

            # Compute gradient εαψηhypotheses
            for dZ2, cache, X in [
                (dZ2_h0, cache_h0, X_h0_sample),
                (dZ2_h1, cache_h1, X_h1_sample),
            ]:
                # ουτ
                gradients["W2"] += np.dot(cache["A1"].T, dZ2)
                gradients["b2"] += np.sum(dZ2, axis=0, keepdims=True)

                # Hidden λαυερ
                dA1 = np.dot(dZ2, parameters["W2"].T)
                dZ1 = dA1 * (1 - np.square(cache["A1"]))
                gradients["W1"] += np.dot(X.T, dZ1)
                gradients["b1"] += np.sum(dZ1, axis=0, keepdims=True)

            optimizer.update(parameters, gradients)

            # Compute cost and error for this sample pair
            sample_cost, sample_error = compute_cost_and_error(
                X_h0_sample, X_h1_sample, parameters
            )
            epoch_costs.append(sample_cost)
            epoch_errors.append(sample_error)

        # Calculate average metrics
        avg_epoch_cost = np.mean(epoch_costs)
        avg_epoch_error = np.mean(epoch_errors)

        cost_history.append(avg_epoch_cost)
        error_history.append(avg_epoch_error)

        # Calculate smoothed

        current_smoothed_cost = get_smoothed_metric(cost_history)
        current_smoothed_error = get_smoothed_metric(error_history)

        smoothed_costs.append(current_smoothed_cost)
        smoothed_errors.append(current_smoothed_error)

        # Check for convergence using smoothed costs
        if len(smoothed_costs) > 1:

            cost_change = abs(smoothed_costs[-1] - smoothed_costs[-2])

            if cost_change < convergence_threshold:
                print(f"Converged at epoch {epoch}")
                print(f"Final smoothed cost: {current_smoothed_cost:.4f}")
                print(f"Final smoothed error: {current_smoothed_error:.4f}")

                return parameters, {
                    "raw_costs": cost_history,
                    "smoothed_costs": smoothed_costs,
                    "errors": error_history,
                    "smoothed_errors": smoothed_errors,
                }

        # Progress logging
        if epoch % 10 == 0:
            print(f"Epoch {epoch}:")
            print(
                f"  Raw Cost: {avg_epoch_cost:.4f}, Smoothed Cost: {current_smoothed_cost:.4f}"
            )
            print(
                f"  Raw Error: {avg_epoch_error:.4f}, Smoothed Error: {current_smoothed_error:.4f}"
            )

    print("Maximum epochs reached without convergence")
    return parameters, {
        "raw_costs": cost_history,
        "smoothed_costs": smoothed_costs,
        "errors": error_history,
        "smoothed_errors": smoothed_errors,
    }
```

Δημιουργώ τη συνάρτηση `evaluate_test_error` για να υπολογίσω το μέσο σφάλμα στα test data. Η διαδικασία περιλαμβάνει:

- Τυχαία επανάληψη (random iterate) πάνω στα `X_test_h0` και `X_test_h1`.
- Υπολογισμό των λαθών βάσει των thresholds που δόθηκαν στην εκφώνηση και στη διάλεξη 6, με βάση τα labels που έχουν οριστεί.
- Επιστροφή του μέσου σφάλματος.

Η `evaluate_test_error` είναι παρόμοια με τη συνάρτηση `evaluate_error`, αλλά έχει υποστεί μικρές προσαρμογές και κάνει merge των δύο υποθέσεων.


```python
# Calculate final test error
def evaluate_test_error(X_test_h0, X_test_h1, parameters, method="cross_entropy"):
    # Combine X_test_h0 and X_test_h1 into a single array
    X_test = np.concatenate((X_test_h0, X_test_h1))

    # Create labels for the combined array
    y_test = np.concatenate((np.zeros(len(X_test_h0)), np.ones(len(X_test_h1))))

    # Shuffle the combined array and labels
    indices = np.arange(len(X_test))
    np.random.shuffle(indices)
    X_test = X_test[indices]
    y_test = y_test[indices]

    # Rest of the code remains the same
    n_samples = len(X_test)
    test_errors = []

    for i in range(n_samples):
        X_sample = X_test[i : i + 1]
        y_sample = y_test[i : i + 1]

        # Forward pass
        cache = forward_pass(X_sample, parameters)

        # Calculate errors (pred should match the labels)
        if method == "cross_entropy":
            pred = sigmoid(cache["Z2"]) > 0.5
        else:  # exponential method
            pred = cache["Z2"] > 0

        error = int(pred)  # false positive rate

        # Calculate accuracy
        accuracy = int(pred == y_sample)

        test_errors.append(accuracy)

    # Calculate average test accuracy
    avg_test_accuracy = np.mean(test_errors)

    return 1 - avg_test_accuracy


test_error_ce = evaluate_test_error(X_test_h0, X_test_h1, params_ce, "cross_entropy")
test_error_exp = evaluate_test_error(X_test_h0, X_test_h1, params_exp, "exponential")

print(f"\nFinal test error rates:")
print(f"Cross-entropy method: {test_error_ce:.4f}")
print(f"Exponential method: {test_error_exp:.4f}")
```

# Problem 2: 

### Απάντηση


Για αυτό το πρόβλημα, καθώς έχω πιο περίπλοκα datasets, θα χρησιμοποιήσω έτοιμα frameworks της Python. Πρακτικά, κάνουμε ακριβώς το ίδιο με το παραπάνω, αλλά με pre-made δεδομένα και training sets, σε πολύ μεγαλύτερη κλίμακα, με τη βοήθεια του TensorFlow.


Αρχικά κάνω τα απαραίτητα imports: την `numpy`, την `mnist`, την `matplotlib.pyplot`, τον ADAM optimizer και την `classification_report` για να φτιάξω το classification report στο τέλος.

Μετά κάνω load τα δεδομένα που θέλω από την MNIST, σύμφωνα με τα docs, και παίρνω μόνο τους αριθμούς 0 και 8. 

Κάνω flatten και normalize όπως ζητά η εκφώνηση.

```python
import tensorflow as tf
from keras import layers, models
from keras import optimizers
import numpy as np
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

Adam = optimizers.Adam

# Load MNIST data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Isolate the numerals 0 and 8 for training and testing
train_filter = np.isin(y_train, [0, 8])
test_filter = np.isin(y_test, [0, 8])


x_train, y_train = x_train[train_filter], y_train[train_filter]
x_test, y_test = x_test[test_filter], y_test[test_filter]


# Convert labels to binary: 0 -> 0, 8 -> 1
y_train = (y_train == 8).astype(int)
y_test = (y_test == 8).astype(int)

# Normalize pixel values to [0, 1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Flatten the 28x28 images into 784-dimensional vectors
x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)
```

Δημιουργώ ένα function για τη δημιουργία ενός neural network.  
Θέτω το activation function σε 'sigmoid' by default, το hidden layer σε 'relu', και δημιουργώ όσα layers ζητά η εκφώνηση.


```python
# Define function to create a neural network
def create_model(activation="sigmoid"):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(784,)))
    model.add(layers.Dense(300, activation="relu"))
    model.add(layers.Dense(1, activation=activation))
    return model
```

Χρησιμοποιώντας το παραπάνω function, δημιουργώ μοντέλο με την μέθοδο cross-entropy loss.  
Κάνω fit (training) με τα δεδομένα που δημιούργησα παραπάνω.  
Η συνάρτηση `fit` σώζει όλα τα απαραίτητα στοιχεία για την εκτίμηση του μοντέλου στο `history`. Από το `history`, παίρνω την `accuracy` και την `val_accuracy` για να υπολογίσω το training error και το validation error αντίστοιχα. Παίρνω επίσης και τα κόστη, τόσο για το training όσο και για το validation, για το cost epoch διάγραμμα.
Τέλος σχεδιάζω τα αποτελέσματα με matplotlib.

```python
# Compile and train model with cross-entropy loss
cross_entropy_model = create_model(activation="sigmoid")
cross_entropy_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=["accuracy"],
)

history_ce = cross_entropy_model.fit(
    x_train, y_train, epochs=14, batch_size=32, validation_split=0.1, verbose=1
)


# Calculate error rate for cross-entropy model
error_rate_ce_train = 1 - np.array(history_ce.history["accuracy"])
error_rate_ce_val = 1 - np.array(history_ce.history["val_accuracy"])

# Plot loss for cross-entropy model
plt.figure(figsize=(10, 5))
plt.plot(history_ce.history["loss"], label="Training Loss (CE)")
plt.plot(history_ce.history["val_loss"], label="Validation Loss (CE)")
plt.title("Cross-Entropy Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Plot error rate for cross-entropy model
plt.figure(figsize=(10, 5))
plt.plot(error_rate_ce_train, label="Training Error Rate (CE)", linestyle="--")
plt.plot(error_rate_ce_val, label="Validation Error Rate (CE)", linestyle="--")
plt.title("Cross-Entropy Model Error Rate")
plt.xlabel("Epochs")
plt.ylabel("Error Rate")
plt.legend()
plt.show()
```

Κάνω ακριβώς το ίδιο, μόνο με το exponential loss model.  
Όμως, αφού η exponential loss function δεν υπάρχει natively, την υλοποιώ εγώ στο μοντέλο. Προσέχω να κάνω set το activation του output σε 'none', για να μην εφαρμοστεί κάποιο activation.


```python
# Compile and train model with custom exponential loss
def exponential_loss(y_true, y_pred):
    return tf.reduce_mean(tf.exp(tf.abs(y_true - y_pred)))


exponential_model = create_model(
    activation=None
)  # No activation for log-likelihood estimation
exponential_model.compile(
    optimizer=Adam(learning_rate=0.001), loss=exponential_loss, metrics=["accuracy"]
)

history_exp = exponential_model.fit(
    x_train, y_train, epochs=14, batch_size=32, validation_split=0.1, verbose=1
)

# Calculate error rate for exponential model
error_rate_exp_train = 1 - np.array(history_exp.history["accuracy"])
error_rate_exp_val = 1 - np.array(history_exp.history["val_accuracy"])

# Plot loss for exponential model
plt.figure(figsize=(10, 5))
plt.plot(history_exp.history["loss"], label="Training Loss (Exp)")
plt.plot(history_exp.history["val_loss"], label="Validation Loss (Exp)")
plt.title("Exponential Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Plot error rate for exponential model
plt.figure(figsize=(10, 5))
plt.plot(error_rate_exp_train, label="Training Error Rate (Exp)", linestyle="--")
plt.plot(error_rate_exp_val, label="Validation Error Rate (Exp)", linestyle="--")
plt.title("Exponential Model Error Rate")
plt.xlabel("Epochs")
plt.ylabel("Error Rate")
plt.legend()
plt.show()
```

Κάνω evaluation μέσω του test set, κοιτώντας απλά το error rate, αλλά και μέσω του `classification_report`,  
 το οποίο δίνει αναλυτικά στατιστικά για την απόδοση του μοντέλου.


```python
# Evaluate on the test set


print("Cross-Entropy Model Performance on Test Set")

ce_eval = cross_entropy_model.evaluate(x_test, y_test, verbose=0)

print(f"Test Loss (CE): {ce_eval[0]}, Test Accuracy (CE): {ce_eval[1]}")

print("\nExponential Model Performance on Test Set")
exp_eval = exponential_model.evaluate(x_test, y_test, verbose=0)
print(f"Test Loss (Exp): {exp_eval[0]}, Test Accuracy (Exp): {exp_eval[1]}")

# Classification report for  δεταιλεδ στατσ
y_pred_ce = (cross_entropy_model.predict(x_test) > 0.5).astype(int)
y_pred_exp = (exponential_model.predict(x_test) > 0).astype(int)

print("\nClassification Report (Cross-Entropy):")
print(classification_report(y_test, y_pred_ce))

print("\nClassification Report (Exponential):")
print(classification_report(y_test, y_pred_exp))
```
